```python
import pandas as pd
import numpy as np
```

```python
customer = pd.read_csv('C:/Users/Rakesh/Downloads/customer.csv')
transaction = pd.read_csv('C:/Users/Rakesh/Downloads/transaction.csv')
day = pd.read_csv('C:/Users/Rakesh/Downloads/day_dim.csv')
active_addr = pd.read_csv('C:/Users/Rakesh/Downloads/active_address.csv')
history_addr = pd.read_csv('C:/Users/Rakesh/Downloads/history_address.csv')
```

```python
day = day.rename(columns={'day_dim_key': 'txn_date_dim_key'})
```

```python
day = day[['txn_date_dim_key', 'month']]
```

```python
merged_transaction_day = pd.merge(transaction, day, on='txn_date_dim_key')
```

```python
acc = pd.merge(customer,merged_transaction_day, on = 'cust_key')
```

```python
acc = pd.merge(active_addr,acc, on = 'customer_id')
```

```python
acc = acc.drop(['address_key', 'eff_start_date_x', 'eff_start_date_y', 'eff_end_date', 
       'transaction_risk', 'txn_key', 'txn_id', 'txn_global_id', 'txn_date_dim_key',
       'txn_original_currency', 'txn_credit_debit_code', 'txn_bene_country_key', 
       'txn_orig_country_key', 'acct_key', 'counterparty_key','acct_currency',
       'acct_open_date_dim_key', 'acct_close_date_dim_key', 'product_key', 'link_type',
       'transaction_type_via', 'insert_date'],axis=1)
```

```python
acc.head()
```

```python
acc.columns
```

```python
month_txn = acc.groupby(['customer_id', 'month'])['txn_amount_eur'].sum()
```

```python
acc = acc.drop_duplicates(subset = ['customer_id'])
```

```python
month_txn
```

```python
acc.head()
```

```python
acc = pd.merge(acc, month_txn, on=['customer_id', 'month'])
```

```python
acc = acc.rename(columns={'txn_amount_eur_y': 'monthly_transaction_total'})
```

```python
acc = acc.drop(['month'],axis=1)
```

```python
acc['has_card'] = np.where(acc['has_card'] == 1, 'Yes', 'No')
acc['has_online_banking'] = np.where(acc['has_online_banking'] == 1, 'Yes', 'No')
```

```python
acc.to_csv('Account_Summary.csv')
```
